Perform data cleaning and exploratory data analysis (EDA) on a data set of your choice, such as titanic dataset from kaggle. explore the relationship between variables and identify patterns and trends in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

test =  pd.read_csv('test.csv')
test
```

Checking Null values

```python
test.isnull().sum()
```

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Cabin'] = test['Cabin'].fillna('Unknown')
```

```python
test.isnull().sum()
```

```python
Le = LabelEncoder()
test['Sex'] = Le.fit_transform(test['Sex'])
test['Embarked'] = Le.fit_transform(test['Embarked'])
```

```python
test
```

```python
train = pd.read_csv('train.csv')
train
```

```python
train.isnull().sum()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Cabin'] = train['Cabin'].fillna('Unknown')
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
train.isnull().sum()
```

```python
train
```

Convert Catrgorical Variables

```python
le = LabelEncoder()
train['Sex'] = le.fit_transform(train['Sex'])
train['Embarked'] = le.fit_transform(train['Embarked'])
```

```python
print("Survived Rate", train['Survived'].mean())
```

Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Survived', data=train)
plt.title("Survived Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(x='Age', hue='Survived', data=train, bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Sex', hue='Survived', data=train, palette='Set1')
plt.title("Sex Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Pclass', hue='Survived', data=train, palette='Set2')
plt.title('Pclass Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Survived", y='Fare', data=train)
plt.title('Fare by Survival')
plt.show()
```

```python
numerical_cols = train.select_dtypes(include=[np.number]).columns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(train[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Embarked", hue="Survived", data=train)
plt.title("Survival by Embarked Port")
plt.show()
```
